## Part 3: Building the Budget Tracker

Now for the real thing. Ask AI to help you write a program that:
- Takes the user's monthly income

- Calculates what's left over
- Warns the user if the budget goes negative

```python
# Write your detailed prompt here:
"""
My Budget Tracker Prompt:
Can you please write a program that:
Takes the user's monthly income
Takes up to 3 expenses
Calculates what's left over
Warns the user if the budget goes negative
"""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
monthly_income = float(input("Enter your monthly income: "))

expenses = []
for i in range(3):
    while True:
        try:
            expense_str = input(f"Enter expense {i+1} (or press Enter to skip): ")
            if expense_str == "":
                break
            expense = float(expense_str)
            if expense < 0:
                print("Expense cannot be negative. Please enter a positive value.")
            else:
                expenses.append(expense)
                break
        except ValueError:
            print("Invalid input. Please enter a number for the expense.")

total_expenses = sum(expenses)
remaining_budget = monthly_income - total_expenses

print(f"\n--- Budget Summary ---")
print(f"Monthly Income: ${monthly_income:,.2f}")
print(f"Total Expenses: ${total_expenses:,.2f}")
print(f"Remaining Budget: ${remaining_budget:,.2f}")

if remaining_budget < 0:
    print("\n⚠️ Warning: Your budget is negative! You are spending more than you earn.")
elif remaining_budget == 0:
    print("\nYour budget is perfectly balanced!")
else:
    print("\nGreat job! You have money left over.")
```

## Part 4: Enhancement Challenge

Pick ONE feature to add with AI's help:
- 🏷️ Expense categories (food, transport, entertainment)
- 📊 Percentage of income spent
- ✅ Input validation (no negative numbers)
- 📝 A detailed budget summary with advice


```python
enhancement_choice = "I chose to add: Expense categories (Food, transport, entertainment)"
enhancement_prompt = """
From this code, can you please add expense categories (Food, transport, entertainment): (CODE)
"""

print(f"🚀 Enhancement: {enhancement_choice}")
```

```python
# ENHANCED BUDGET TRACKER
monthly_income = float(input("Enter your monthly income: "))

categorized_expenses = {
    'Food': [],
    'Transport': [],
    'Entertainment': [],
    'Other': []
}

for i in range(3):
    while True:
        try:
            expense_str = input(f"Enter expense {i+1} (or press Enter to skip): ")
            if expense_str == "":
                break
            expense = float(expense_str)
            if expense < 0:
                print("Expense cannot be negative. Please enter a positive value.")
            else:
                # Get category
                while True:
                    category = input(f"Enter category for expense {i+1} (Food, Transport, Entertainment, Other): ").strip().capitalize()
                    if category in categorized_expenses:
                        categorized_expenses[category].append(expense)
                        break
                    else:
                        print("Invalid category. Please choose from Food, Transport, Entertainment, Other.")
                break
        except ValueError:
            print("Invalid input. Please enter a number for the expense.")

total_expenses = 0
for category in categorized_expenses:
    total_expenses += sum(categorized_expenses[category])

remaining_budget = monthly_income - total_expenses

print(f"\n--- Budget Summary ---")
print(f"Monthly Income: ${monthly_income:,.2f}")

print("\n--- Expenses by Category ---")
for category, expenses_list in categorized_expenses.items():
    category_total = sum(expenses_list)
    print(f"{category}: ${category_total:,.2f}")

print(f"\nTotal Expenses: ${total_expenses:,.2f}")
print(f"Remaining Budget: ${remaining_budget:,.2f}")

if remaining_budget < 0:
    print("\n⚠️ Warning: Your budget is negative! You are spending more than you earn.")
elif remaining_budget == 0:
    print("\nYour budget is perfectly balanced!")
else:
    print("\nGreat job! You have money left over.")
```
